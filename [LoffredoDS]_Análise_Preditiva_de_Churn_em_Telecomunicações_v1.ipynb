# O que  *Churn* e por que ele  importante?

O termo em ingls *churn* no possui uma traduo para o portugus, mas ele  bem simples de ser compreendido. *Churn*  uma mtrica que mostra a taxa de evaso de clientes em uma empresa. Ou, em empresas que fornecem servios por assinatura como  o caso do Spotify, Netflix e Amazon Prime, essa taxa indica a quantidade de assinaturas canceladas.
</br>
</br>

Saber a taxa de *churn*  importante pois busca-se o menor valor possvel e, com a noo dela ao longo do tempo pode-se validar estratgias de marketing, valores cobrados, etc. Sempre com o objetivo de maximizar o lucro da empresa e evitar prejuzos.

<br/>

<br/>



Seu clculo  dado pela diviso da quantidade de clientes que foram perdidos pelo total de clientes ativos.
</br>
</br>

$$
Taxa\space de\space Churn = \frac{quantidade\space de\space clientes\space perdidos}{total\space de\space clientes\space ativos\space}
$$
</br>
</br>

Isso quer dizer que se uma empresa tinha 1000 clientes e perdeu 10, sua taxa de *churn*  de 1%.
</br>
</br>

O *Churn Prediction* ou, a previso de *churn*,  extremamente importante e pode ser feita com o uso de tcnicas de aprendizado de mquina e com dados passados de rotatividade de clientes. Com isso, ser possvel, em conjunto com o setor de Marketing, realizar aes para prevenir a perda desses clientes apontados como possveis *churns*. Essa converso de clientes pode ser muito benfica para a empresa, uma vez que o custo de aquisio de novos clientes, dados pela mtrica CAC, chega a ser 15x mais alto do que o valor gasto em campanhas para reter um cliente.

## 3. Tratar as variveis categricas

Na sequncia, vamos tratar as variveis categricas. Para as variveis que possuem 2 tipos de entradas, por exemplo sim/no, ser utilizado o *Label Encoding* uma vez que ele converte os dados em numricos e, como possumos 2 entradas, eles se tornaro booleanos pois ficaro com os dados em 0 ou 1.
Quanto aos atributos categricos que no se encaixam no processamento acima, ou seja, possuem 3 ou mais entradas diferentes, eles sero processados com a utilizao de variveis *dummy*.

```python
# identificar as variveis com 2 labels
binary_var = df_clean.nunique()[df_clean.nunique() == 2].keys().tolist()

# distinguir os atributos categricos dos numricos
## identificar os atributos numricos
num_var = [col for col in df_clean.select_dtypes(['int', 'float']).columns.tolist() if col not in binary_var]

## identificar os atributos categricos sendo aqueles que no so os numricos, e nem so os categricos j tratados na etapa anterior
cat_var = [col for col in df_clean.columns.tolist() if col not in binary_var + num_var]

# criar uma cpia do data frame
df_proc = df_clean.copy()

# aplicar o Label Encoding
le = LabelEncoder()
for i in binary_var:
    df_proc[i] = le.fit_transform(df_proc[i])

# aplicar variveis dummy
df_proc = pd.get_dummies(df_proc, columns=cat_var)

# checar alteraes no data frame
df_proc.head()
```

# Criao de modelos de *Machine Learning*

## Funo de `val_model`

Para iniciar essa etapa, em primeiro lugar vou construir uma funo denominada `val_model` com a construo de um *pipeline* (fluxo de trabalho) que ir padronizar os dados com o  *Standard Scaler*, classific-los (de acordo com o que for inserido na funo pelo parmetro `clf`) e fazer a avaliao do modelo por meio da validao cruzada, levando em considerao o valor de *Recall* obtido. O resultado final dessa funo ser mostrar o valor mdio de *Recall* encontrado na validao cruzada.

```python
def val_model(X, y, clf, quite=False):
    """
    Realiza cross-validation com os dados de treino para um determinado modelo.

    # Arguments
        X: Data Frame, contm as variveis independentes.
        y: Series, vetor contendo a varivel alvo.
        clf: modelo classificador do scikit-learn.
        quite: bool, indicando se a funo deve imprimir os resultados ou no.

    # Returns
        float, mdia dos scores da cross-validation.
    """

    # converter variveis em arrays
    X = np.array(X)
    y = np.array(y)

    # criao de fluxo de trabalho
    ## 1. padronizar os dados com o StandardScaler
    ## 2. classificar os dados
    pipeline = make_pipeline(StandardScaler(), clf)

    # avaliao do modelo por validao cruzada
    ## de acordo com o valor de Recall
    scores = cross_val_score(pipeline, X, y, scoring='recall')

    # mostrar mdia do valor de Recall e desvio padro do modelo
    if quite == False:
        print("Recall: {:.2f} (+/- {:.2f})".format(scores.mean(), scores.std()))

    # retornar a mdia dos valores de Recall obtidos na validao cruzada
    return scores.mean()
```

Ademais, esse modelo base no incluir ajustes de hiperparmetros, nem *feature engineering*, to pouco os dados sero balanceados. Isso para que tenhamos o valor base mais bruto possvel para tambm checarmos o quanto  possvel melhorar o resultado.

## Mtrica de Avaliao de Performance

Para avaliar o desempenho dos modelos a mtrica que ser utilizada  o valor de *Recall*. Isso porque quando se tem dados desbalanceados, como  o caso do conjunto em estudo, ainda que ele j tenha sido tratado e balanceado, a acurcia no  uma boa mtrica de avaliao. O motivo disso se deve ao fato de que se pode obter um resultado de acurcia muito alto, entretanto, na identificao de *churn*, a deteco ter um resultado muito baixo, o que no nos leva ao nosso objetivo.
<br/>
<br/>

O *Recall*  a mtrica que nos fornece a melhor medida para o problema especfico em estudo. Isso se deve porque, no caso de *churn*, os Falsos Negativos so mais prejudiciais para uma empresa do que os Falsos Positivos. Em outras palavras,  melhor que o modelo erre ao dizer que um cliente  *churn*, quando na realidade ele no seja. Do que erre ao apontar um cliente como no sendo *churn*, quando na verdade ele , trazendo prejuzos para os negcios.

Tendo isso em vista, busca-se uma alta taxa de *Recall*.

O *Recall* olha para todas os *churns* e visa responder: o quanto o modelo acerta? O resultado ser um valor entre 0 e 1, sendo que, quanto mais prximo de 1, melhor, pois indica uma baixa taxa de Falsos Negativos.
<br/>
<br/>

Seu clculo  dado por:
<br/>
<br/>

$${\text{Recall: }}\frac{\text{Verdadeiro Positivo}} {\text{Verdadeiro Positivo+ Falso Negativo}}$$
<br/>
<br/>

Ainda pensando na finalidade desse estudo, outra mtrica de avaliao de modelos de classificao que pode ser utilizada  a AUC - *Area Under the Curve*, traduzindo  a Area abaixo da curva. Ela deriva do ROC - *Receiver Operating Characteristic* - e indica para ns o quanto o modelo consegue distinguir entre duas coisas. No nosso exemplo, entre um cliente *churn* e no-*churn*. O valor dado por AUC varia entre 0 e 1, sendo que, quanto mais prximo de 1, indica que melhor  o modelo.

# XGBoost: Otimizao de Hiperparmetros

Por conta do problema que estamos tentando revolver (ver [Objetivo Geral](#scrollTo=5rSx8OkFXz67)), bem como, pelo desempenho dos algoritmos na sesso anterior. O XGboost possui um potencial de melhorar muito seu resultado por meio de ajustes nos seus hiperparmetros. O que, comparativamente com a maior parte dos algoritmos acima,  uma quantidade maior e que, portanto, d essa probabilidade maior tambm de aprimorar a sua performance.

Em razo da quantidade de hiperparmetros que precisam ser ajustados,  necessrio encontrar o melhor valor para cada um deles em processos distintos. Para iniciar, se definiu a taxa de aprendizado em 0.1. A recomendao  de seja entre 0.05 e 0.3, de acordo com o problema a ser resolvido.

Vamos fazer a busca para o melhor parmetro para `n_estimators` que  a quantidade de rvores de deciso que deve conter no algoritmo.

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
xgb = XGBClassifier(learning_rate=0.1)

# definir dicionrio para descobrir a quantidade ideal de rvores
## em um range de 0 a 500 com incremento de 50
param_grid = {'n_estimators':range(0,500,50)}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros de rvores do param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'n_estimators'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

O melhor resultado encontrado foi com 50 rvores de deciso. Com esse parmetro, como o incremento definido foi de 50, podemos refinar a buscar em torno desse valor, com passos menores:

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
xgb = XGBClassifier(learning_rate=0.1)

# definir dicionrio para descobrir a quantidade ideal de rvores
## em um range de 0 a 100 com incremento de 10
param_grid = {'n_estimators':range(0,100,10)}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros de rvores do param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'n_estimators'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Com essa nova busca, o melhor valor encontrado para a quantidade de rvores foi de 70. Apenas com esse ajuste melhoramos ainda mais o desempenho desse algoritmo. Ou seja, com apenas 1 *tunning* conseguimos com que o XGBoost ficasse agora na 3 colocao na lista de modelos gerada na sesso anterior. Temos outros 4 *tunnings* pela frente, logo, parece promissor!

Com o melhor parmetro para `n_estimators`, podemos defini-lo na atribuio do classificador e prosseguir para determinar os melhores valores para `max_depth`, que determina a profundidade das rvores de deciso e, para o `min_child_weight`, ou seja, o peso mnimo do n para que seja criado um novo n.

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
## n_estimators: 70
xgb = XGBClassifier(learning_rate=0.1, n_estimators=70)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 1 a 7 com incremento de 1
## o peso mnimo do n em um range de 1 a 4 com incremento de 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros de rvores do param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'max_depth' e 'min_child_weight'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'max_depth' e 'min_child_weight'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Os valores obtidos para ambos so os menores possveis, portanto, no h como tentar obter um resultado ainda melhor.

Note que, at esse passo, agora j superamos o melhor valor de *Recall* encontrado anteriormente de 0.8017, dado ao modelo de Regresso Logstica.

Temos mais 2 parmetros para ajustar!

Podemos inserir os novos parmetros encontrados e partir para encontrar o prximo: `gamma`. Esse parmetro define a complexidade das rvores no modelo.

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
## n_estimators: 70
## max_depth: 1
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=70, max_depth=1, min_child_weight=1)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 0.0 a 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros de rvores do param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'gamma'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'gamma'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Mais uma vez, o melhor valor encontrado foi o valor mnimo, ento, prosseguimos inserindo esse valor na atribuio do modelo de classificao e buscando o melhor valor para a taxa de aprendizado (`learning_rate`).

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
## n_estimators: 70
## max_depth: 1
## min_child_weight: 1
## gamma = 0.0
xgb = XGBClassifier(n_estimators=70, max_depth=1, min_child_weight=1, gamma=0.0)

# definir dicionrio para descobrir a taxa de aprendizado ideal
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros de rvores do param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'learning_rate'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parmetro encontrado para 'learning_rate'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

```python
# instanciar o modelo XGBoost
## taxa de aprendizado: 0.1
## n_estimators: 30
xgb = XGBClassifier(learning_rate=0.1, n_estimators=30)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 1 a 7 com incremento de 1
## o peso mnimo do n em um range de 1 a 4 com incremento de 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros configurados em param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca dos melhores parmetros para 'max_depth' e 'min_child_weight'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'max_depth' e 'min_child_weight'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Podemos inserir os novos parmetros encontrados e partir para encontrar o prximo: `gamma`. Esse parmetro define a complexidade das rvores no modelo.

```python
# instanciar o modelo XGBoost
## taxa de aprendizado: 0.1
## n_estimators: 30
## max_depth: 1
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=30, max_depth=1, min_child_weight=1)

# definir dicionrio para descobrir:
## a profundidade ideal em um range de 0.0 a 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros configurados em param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'gamma'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'gamma'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Mais uma vez, o melhor valor encontrado foi o valor mnimo, ento, prosseguimos inserindo esse valor na atribuio do modelo de classificao e buscando o melhor valor para a taxa de aprendizado (`learning_rate`).

```python
# instanciar o modelo XGBoost
## taxa de aprendizado: 0.1
## n_estimators: 30
## max_depth: 1
## min_child_weight: 1
## gamma = 0.0
xgb = XGBClassifier(n_estimators=30, max_depth=1, min_child_weight=1, gamma=0.0)

# definir dicionrio para descobrir a taxa de aprendizado ideal
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# configurar validao cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configurao da busca de combinaes cruzadas com o classificador XGBoost
## parmetros configurados em param_grid
## scoring: mtodo de avaliao por Recall
## n_jobs=1 para busca em paralelo (usando todos os ncleos disponveis)
## cv: estratgia de validao
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parmetro para 'learning_rate'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parmetro encontrado para 'learning_rate'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Dos valores testados, o que representou melhor resultado foi com uma taxa de aprendizado de 0.01. Poderamos testar valores mais baixos, porm, quanto menor for esse valor maior  o risco de *overfitting* do modelo. Por isso, vou manter esse valor.
