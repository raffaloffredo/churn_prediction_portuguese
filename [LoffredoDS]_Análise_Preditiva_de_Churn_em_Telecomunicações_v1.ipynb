# O que é *Churn* e por que ele é importante?

O termo em inglês *churn* não possui uma tradução para o português, mas ele é bem simples de ser compreendido. *Churn* é uma métrica que mostra a taxa de evasão de clientes em uma empresa. Ou, em empresas que fornecem serviços por assinatura como é o caso do Spotify, Netflix e Amazon Prime, essa taxa indica a quantidade de assinaturas canceladas.
</br>
</br>

Saber a taxa de *churn* é importante pois busca-se o menor valor possível e, com a noção dela ao longo do tempo pode-se validar estratégias de marketing, valores cobrados, etc. Sempre com o objetivo de maximizar o lucro da empresa e evitar prejuízos.

<br/>

<br/>



Seu cálculo é dado pela divisão da quantidade de clientes que foram perdidos pelo total de clientes ativos.
</br>
</br>

$$
Taxa\space de\space Churn = \frac{quantidade\space de\space clientes\space perdidos}{total\space de\space clientes\space ativos\space}
$$
</br>
</br>

Isso quer dizer que se uma empresa tinha 1000 clientes e perdeu 10, sua taxa de *churn* é de 1%.
</br>
</br>

O *Churn Prediction* ou, a previsão de *churn*, é extremamente importante e pode ser feita com o uso de técnicas de aprendizado de máquina e com dados passados de rotatividade de clientes. Com isso, será possível, em conjunto com o setor de Marketing, realizar ações para prevenir a perda desses clientes apontados como possíveis *churns*. Essa conversão de clientes pode ser muito benéfica para a empresa, uma vez que o custo de aquisição de novos clientes, dados pela métrica CAC, chega a ser 15x mais alto do que o valor gasto em campanhas para reter um cliente.

## 3. Tratar as variáveis categóricas

Na sequência, vamos tratar as variáveis categóricas. Para as variáveis que possuem 2 tipos de entradas, por exemplo sim/não, será utilizado o *Label Encoding* uma vez que ele converte os dados em numéricos e, como possuímos 2 entradas, eles se tornarão booleanos pois ficarão com os dados em 0 ou 1.
Quanto aos atributos categóricos que não se encaixam no processamento acima, ou seja, possuem 3 ou mais entradas diferentes, eles serão processados com a utilização de variáveis *dummy*.

```python
# identificar as variáveis com 2 labels
binary_var = df_clean.nunique()[df_clean.nunique() == 2].keys().tolist()

# distinguir os atributos categóricos dos numéricos
## identificar os atributos numéricos
num_var = [col for col in df_clean.select_dtypes(['int', 'float']).columns.tolist() if col not in binary_var]

## identificar os atributos categóricos sendo aqueles que não são os numéricos, e nem são os categóricos já tratados na etapa anterior
cat_var = [col for col in df_clean.columns.tolist() if col not in binary_var + num_var]

# criar uma cópia do data frame
df_proc = df_clean.copy()

# aplicar o Label Encoding
le = LabelEncoder()
for i in binary_var:
    df_proc[i] = le.fit_transform(df_proc[i])

# aplicar variáveis dummy
df_proc = pd.get_dummies(df_proc, columns=cat_var)

# checar alterações no data frame
df_proc.head()
```

## Função de `val_model`

Para iniciar essa etapa, em primeiro lugar vou construir uma função denominada `val_model` com a construção de um *pipeline* (fluxo de trabalho) que irá padronizar os dados com o  *Standard Scaler*, classificá-los (de acordo com o que for inserido na função pelo parâmetro `clf`) e fazer a avaliação do modelo por meio da validação cruzada, levando em consideração o valor de *Recall* obtido. O resultado final dessa função será mostrar o valor médio de *Recall* encontrado na validação cruzada.

```python
def val_model(X, y, clf, quite=False):
    """
    Realiza cross-validation com os dados de treino para um determinado modelo.

    # Arguments
        X: Data Frame, contém as variáveis independentes.
        y: Series, vetor contendo a variável alvo.
        clf: modelo classificador do scikit-learn.
        quite: bool, indicando se a função deve imprimir os resultados ou não.

    # Returns
        float, média dos scores da cross-validation.
    """

    # converter variáveis em arrays
    X = np.array(X)
    y = np.array(y)

    # criação de fluxo de trabalho
    ## 1. padronizar os dados com o StandardScaler
    ## 2. classificar os dados
    pipeline = make_pipeline(StandardScaler(), clf)

    # avaliação do modelo por validação cruzada
    ## de acordo com o valor de Recall
    scores = cross_val_score(pipeline, X, y, scoring='recall')

    # mostrar média do valor de Recall e desvio padrão do modelo
    if quite == False:
        print("Recall: {:.2f} (+/- {:.2f})".format(scores.mean(), scores.std()))

    # retornar a média dos valores de Recall obtidos na validação cruzada
    return scores.mean()
```

## Balancear o conjunto de dados

Como visto, o conjunto de dados está desbalanceado e é necessário rbalancear os dados para evitar a criação de um modelo que tenha baixa performance na identificação de *churn*, bem como evitar o *overfitting*. Dessa forma criamos um bom modelo de *machine learning* livre de viés.

O método utilizado para ser aplicado nesse caso é o ***Undersampling***, que reduz a classe majoritária, excluindo esses dados de forma aleatória. Assim, as características da classe minoritária, que nesse caso são os dados de *churn*, são preservados. E esses são os dados mais importantes para a solução do nosso problema.

Essa técnica é aplicada apenas no conjunto de treino para que as características do conjunto de testes não sejam desconfiguradas.

```python
# instanciar modelo de padronização
scaler = StandardScaler().fit(X_train)

# aplicar padronização nos dados de treino
## apenas nas variáveis independentes
X_train = scaler.transform(X_train)

# instanciar modelo de undersampling
## random_state para que o resultado seja replicável
rus = RandomUnderSampler(random_state=43)

# aplicar undersampling nos dados de treino
X_train_rus, y_train_rus = rus.fit_resample(X_train, y_train)

# verificar o balanceamento das classes
print(pd.Series(y_train_rus).value_counts())
```

## Métrica de Avaliação de Performance

Para avaliar o desempenho dos modelos a métrica que será utilizada é o valor de *Recall*. Isso porque quando se tem dados desbalanceados, como é o caso do conjunto em estudo, ainda que ele já tenha sido tratado e balanceado, a acurácia não é uma boa métrica de avaliação. O motivo disso se deve ao fato de que se pode obter um resultado de acurácia muito alto, entretanto, na identificação de *churn*, a detecção ter um resultado muito baixo, o que não nos leva ao nosso objetivo.
<br/>
<br/>

O *Recall* é a métrica que nos fornece a melhor medida para o problema específico em estudo. Isso se deve porque, no caso de *churn*, os Falsos Negativos são mais prejudiciais para uma empresa do que os Falsos Positivos. Em outras palavras, é melhor que o modelo erre ao dizer que um cliente é *churn*, quando na realidade ele não seja. Do que erre ao apontar um cliente como não sendo *churn*, quando na verdade ele é, trazendo prejuízos para os negócios.

Tendo isso em vista, busca-se uma alta taxa de *Recall*.

O *Recall* olha para todas os *churns* e visa responder: o quanto o modelo acerta? O resultado será um valor entre 0 e 1, sendo que, quanto mais próximo de 1, melhor, pois indica uma baixa taxa de Falsos Negativos.
<br/>
<br/>

Seu cálculo é dado por:
<br/>
<br/>

$${\text{Recall: }}\frac{\text{Verdadeiro Positivo}} {\text{Verdadeiro Positivo+ Falso Negativo}}$$
<br/>
<br/>

Ainda pensando na finalidade desse estudo, outra métrica de avaliação de modelos de classificação que pode ser utilizada é a AUC - *Area Under the Curve*, traduzindo é a Area abaixo da curva. Ela deriva do ROC - *Receiver Operating Characteristic* - e indica para nós o quanto o modelo consegue distinguir entre duas coisas. No nosso exemplo, entre um cliente *churn* e não-*churn*. O valor dado por AUC varia entre 0 e 1, sendo que, quanto mais próximo de 1, indica que melhor é o modelo.
<br/>
<br/>

# XGBoost: Otimização de Hiperparâmetros

Por conta do problema que estamos tentando revolver (ver [Objetivo Geral](#scrollTo=5rSx8OkFXz67)), bem como, pelo desempenho dos algoritmos na sessão anterior. O XGboost possui um potencial de melhorar muito seu resultado por meio de ajustes nos seus hiperparâmetros. O que, comparativamente com a maior parte dos algoritmos acima, é uma quantidade maior e que, portanto, dá essa probabilidade maior também de aprimorar a sua performance.

Em razão da quantidade de hiperparâmetros que precisam ser ajustados, é necessário encontrar o melhor valor para cada um deles em processos distintos. Para iniciar, se definiu a taxa de aprendizado em 0.1. A recomendação é de seja entre 0.05 e 0.3, de acordo com o problema a ser resolvido.

Vamos fazer a busca para o melhor parâmetro para `n_estimators` que é a quantidade de árvores de decisão que deve conter no algoritmo.

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
xgb = XGBClassifier(learning_rate=0.1)

# definir dicionário para descobrir a quantidade ideal de árvores
## em um range de 0 a 500 com incremento de 50
param_grid = {'n_estimators':range(0,500,50)}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros de árvores do param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parâmetro para 'n_estimators'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parâmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

O melhor resultado encontrado foi com 50 árvores de decisão. Com esse parâmetro, como o incremento definido foi de 50, podemos refinar a buscar em torno desse valor, com passos menores:

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
xgb = XGBClassifier(learning_rate=0.1)

# definir dicionário para descobrir a quantidade ideal de árvores
## em um range de 0 a 100 com incremento de 10
param_grid = {'n_estimators':range(0,100,10)}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros de árvores do param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parâmetro para 'n_estimators'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parâmetro encontrado para 'n_estimators'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Com essa nova busca, o melhor valor encontrado para a quantidade de árvores foi de 70. Apenas com esse ajuste melhoramos ainda mais o desempenho desse algoritmo. Ou seja, com apenas 1 *tunning* conseguimos com que o XGBoost ficasse agora na 3ª colocação na lista de modelos gerada na sessão anterior. Temos outros 4 *tunnings* pela frente, logo, parece promissor!

Com o melhor parâmetro para `n_estimators`, podemos defini-lo na atribuição do classificador e prosseguir para determinar os melhores valores para `max_depth`, que determina a profundidade das árvores de decisão e, para o `min_child_weight`, ou seja, o peso mínimo do nó para que seja criado um novo nó.

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
## n_estimators: 70
xgb = XGBClassifier(learning_rate=0.1, n_estimators=70)

# definir dicionário para descobrir:
## a profundidade ideal em um range de 1 a 7 com incremento de 1
## o peso mínimo do nó em um range de 1 a 4 com incremento de 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros de árvores do param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parâmetro para 'max_depth' e 'min_child_weight'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parâmetro encontrado para 'max_depth' e 'min_child_weight'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Os valores obtidos para ambos são os menores possíveis, portanto, não há como tentar obter um resultado ainda melhor.

Note que, até esse passo, agora já superamos o melhor valor de *Recall* encontrado anteriormente de 0.8017, dado ao modelo de Regressão Logística.

Temos mais 2 parâmetros para ajustar!

Podemos inserir os novos parâmetros encontrados e partir para encontrar o próximo: `gamma`. Esse parâmetro define a complexidade das árvores no modelo.

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
## n_estimators: 70
## max_depth: 1
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=70, max_depth=1, min_child_weight=1)

# definir dicionário para descobrir:
## a profundidade ideal em um range de 0.0 a 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros de árvores do param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parâmetro para 'gamma'
grid_result= grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parâmetro encontrado para 'gamma'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Mais uma vez, o melhor valor encontrado foi o valor mínimo, então, prosseguimos inserindo esse valor na atribuição do modelo de classificação e buscando o melhor valor para a taxa de aprendizado (`learning_rate`).

```python
# instanciar o modelo XGBoost
## definir a taxa de aprendizado em 0.1
## n_estimators: 70
## max_depth: 1
## min_child_weight: 1
## gamma = 0.0
xgb = XGBClassifier(n_estimators=70, max_depth=1, min_child_weight=1, gamma=0.0)

# definir dicionário para descobrir a taxa de aprendizado ideal
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros de árvores do param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parâmetro para 'learning_rate'
grid_result = grid_search.fit(X_train_rus, y_train_rus)

# imprimir o melhor parâmetro encontrado para 'learning_rate'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Dos valores testados, o que apresentou melhor resultado foi com uma taxa de aprendizado de 0.01. Poderíamos testar valores mais baixos, porém, quanto menor for esse valor maior é o risco de *overfitting* do modelo. Por isso, vou manter esse valor encontrado.

```python
# instanciar o modelo XGBoost
## taxa de aprendizado: 0.1
## n_estimators: 30
xgb = XGBClassifier(learning_rate=0.1, n_estimators=30)

# definir dicionário para descobrir:
## a profundidade ideal em um range de 1 a 7 com incremento de 1
## o peso mínimo do nó em um range de 1 a 4 com incremento de 1
param_grid = {'max_depth':range(1,8,1),
              'min_child_weight':range(1,5,1)}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros configurados em param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca dos melhores parâmetros para 'max_depth' e 'min_child_weight'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parâmetro encontrado para 'max_depth' e 'min_child_weight'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Podemos inserir os novos parâmetros encontrados e partir para encontrar o próximo: `gamma`. Esse parâmetro define a complexidade das árvores no modelo.

```python
# instanciar o modelo XGBoost
## taxa de aprendizado: 0.1
## n_estimators: 30
## max_depth: 1
## min_child_weight: 1
xgb = XGBClassifier(learning_rate=0.1, n_estimators=30, max_depth=1, min_child_weight=1)

# definir dicionário para descobrir:
## a profundidade ideal em um range de 0.0 a 0.4
param_grid = {'gamma':[i/10.0 for i in range(0,5)]}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros configurados em param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parâmetro para 'gamma'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parâmetro encontrado para 'gamma'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Mais uma vez, o melhor valor encontrado foi o valor mínimo, então, prosseguimos inserindo esse valor na atribuição do modelo de classificação e buscando o melhor valor para a taxa de aprendizado (`learning_rate`).

```python
# instanciar o modelo XGBoost
## taxa de aprendizado: 0.1
## n_estimators: 30
## max_depth: 1
## min_child_weight: 1
## gamma = 0.0
xgb = XGBClassifier(n_estimators=30, max_depth=1, min_child_weight=1, gamma=0.0)

# definir dicionário para descobrir a taxa de aprendizado ideal
param_grid = {'learning_rate':[0.01, 0.05, 0.1, 0.2]}

# configurar validação cruzada com 10 dobras estratificadas
## shuffle=True para embaralhar os dados antes de dividir
kfold = StratifiedKFold(n_splits=10, shuffle=True)

# configuração da busca de combinações cruzadas com o classificador XGBoost
## parâmetros configurados em param_grid
## scoring: método de avaliação por Recall
## n_jobs=1 para busca em paralelo (usando todos os núcleos disponíveis)
## cv: estratégia de validação
grid_search = GridSearchCV(xgb, param_grid, scoring="recall", n_jobs=-1, cv=kfold)

# realizar a busca do melhor parâmetro para 'learning_rate'
grid_result = grid_search.fit(X_fe_train_rus, y_fe_train_rus)

# imprimir o melhor parâmetro encontrado para 'learning_rate'
print("Melhor: {:.4f} para {}".format(grid_result.best_score_, grid_result.best_params_))
```

Dos valores testados, o que representou melhor resultado foi com uma taxa de aprendizado de 0.01. Poderíamos testar valores mais baixos, porém, quanto menor for esse valor maior é o risco de *overfitting* do modelo. Por isso, vou manter esse valor.

Ademais, vou usar o *dataset* original para fazer esse cálculo e, utilizar a métrica de *Recall* obtida a partir do conjunto de testes, que foi de 0.8887, ou seja, de 100% do total de *churn*, o modelo poderia acertar cerca de 88.87% dos casos de *churn*.

```python
# calcular receita total do período
receita_total = (df_clean.MonthlyCharges * df_clean.tenure).sum()

print('Receita Total: {:.2f}'.format(receita_total))
```

Temos que a receita total do período foi de aproximadamente 16 milhões.

```python
# fazer cópia dos atributos necessários
clientes_churn = (df_clean[["tenure", "MonthlyCharges", "Churn"]]).loc[df_clean.Churn == "Yes"]

# imprimir primeiros registros
clientes_churn.head()
```
